```python
import numpy as np
x = np.linspace(0, 10, 50)
y = x

plt.title("Linear graph")
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(x, y, "r.")
```

```python
x = np.linspace(12, 20, 14)
y1 = x
y2 = [i**3 for i in x]
plt.title("y1 = x, y2 = x^2")
plt.xlabel("x")
plt.ylabel("y1, y2")
plt.grid()
plt.plot(x, y1, x, y2)
```

```python
x = np.linspace(0, 25, 30)
y1 = x
y2 = [i**2 for i in x]

plt.figure(figsize=(8,8))
plt.subplot(2, 1, 1)
plt.plot(x, y1, "r.")
plt.title("Functions")
plt.ylabel("y1", fontsize=14)
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(x, y2, "g--")
plt.xlabel("x", fontsize=14)
plt.ylabel("y2", fontsize=14)
plt.grid(True)
```

```python
x = [20, 7, 16, 30, 22]
y1 = [2, 4, 6, 8, 10]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]

plt.plot(x, y1, '-', x, y2, '--', x, y3, '-.', x, y4, ':')
```
